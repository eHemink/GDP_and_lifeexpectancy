```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read in data
df = pd.read_csv("all_data.csv")
print(df.head())
```

## Visualizing Data

```python
# Unique countries and years 
print(df.Country.unique())
print(df.Year.unique())
```

```python
# Has life expectancy increased over time?
ax = sns.scatterplot(x= 'Year', y='Life expectancy at birth (years)', data=df, hue='Country')
sns.move_legend(ax, "upper right", bbox_to_anchor = (1.48,1))
plt.show()
plt.clf()

# Life expectancy has generally been increasing over time in each country. In Zimbabwe, it has increased significantly since around 2008. 
```

```python
# Subsetting data by country and finding average life expectancy in each country
chile_data = df[df.Country == 'Chile']
print("Average life expectancy in Chile: " + str(chile_data['Life expectancy at birth (years)'].mean()))

china_data = df[df.Country == 'China']
print("Average life expectancy in China: " + str(china_data['Life expectancy at birth (years)'].mean()))

germany_data = df[df.Country == 'Germany']
print("Average life expectancy in Germany: " + str(germany_data['Life expectancy at birth (years)'].mean()))

mexico_data = df[df.Country == 'Mexico']
print("Average life expectancy in Mexico: " + str(mexico_data['Life expectancy at birth (years)'].mean()))

usa_data = df[df.Country == 'United States of America']
print("Average life expectancy in USA: " + str(usa_data['Life expectancy at birth (years)'].mean()))

zimbabwe_data = df[df.Country == 'Zimbabwe']
print("Average life expectancy in Zimbabwe: " + str(zimbabwe_data['Life expectancy at birth (years)'].mean()))
```

```python
# Has GDP increased over time in each country? 
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(8, 6))
fig.tight_layout(h_pad = 4)

plt.subplot(2,3,1)
plt.scatter('Year', 'GDP', data = chile_data, color = 'blue')
plt.title('GDP in Chile Over Time')
plt.xlabel('Year')
plt.ylabel('GDP')

plt.subplot(2,3,2)
plt.scatter('Year', 'GDP', data = china_data, color = 'green')
plt.title('GDP in China Over Time')
plt.xlabel('Year')
plt.ylabel('GDP')

plt.subplot(2,3,3)
plt.scatter('Year', 'GDP', data = germany_data, color = 'orange')
plt.title('GDP in Germany Over Time')
plt.xlabel('Year')
plt.ylabel('GDP')

plt.subplot(2,3,4)
plt.scatter('Year', 'GDP', data = mexico_data, color = 'magenta')
plt.title('GDP in Mexico Over Time')
plt.xlabel('Year')
plt.ylabel('GDP')

plt.subplot(2,3,5)
plt.scatter('Year', 'GDP', data = usa_data, color = 'black')
plt.title('GDP in USA Over Time')
plt.xlabel('Year')
plt.ylabel('GDP')

plt.subplot(2,3,6)
plt.scatter('Year', 'GDP', data = zimbabwe_data, color = 'red')
plt.title('GDP in Zimbabwe Over Time')
plt.xlabel('Year')
plt.ylabel('GDP')

plt.show()
plt.clf()

# GDP has generally increased over time in each country. In Zimbabwe around 2007/2008, there was hyperinflation which accounts for the very low GDP.
```

```python
# Is there a correlation between GDP and life expectancy? 
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(8, 6))
fig.tight_layout(h_pad = 4)
plt.suptitle('GDP vs Life Expectancy')
fig.subplots_adjust(top=0.9)

plt.subplot(2,3,1)
plt.scatter('Life expectancy at birth (years)', 'GDP', data = chile_data, color = 'blue')
plt.title('Chile')
plt.xlabel('Life Expectancy')
plt.ylabel('GDP')

plt.subplot(2,3,2)
plt.scatter('Life expectancy at birth (years)', 'GDP', data = china_data, color = 'green')
plt.title('China')
plt.xlabel('Life Expectancy')
plt.ylabel('GDP')

plt.subplot(2,3,3)
plt.scatter('Life expectancy at birth (years)', 'GDP', data = germany_data, color = 'orange')
plt.title('Germany')
plt.xlabel('Life Expectancy')
plt.ylabel('GDP')

plt.subplot(2,3,4)
plt.scatter('Life expectancy at birth (years)', 'GDP', data = mexico_data, color = 'magenta')
plt.title('Mexico')
plt.xlabel('Life Expectancy')
plt.ylabel('GDP')

plt.subplot(2,3,5)
plt.scatter('Life expectancy at birth (years)', 'GDP', data = usa_data, color = 'black')
plt.title('USA')
plt.xlabel('Life Expectancy')
plt.ylabel('GDP')

plt.subplot(2,3,6)
plt.scatter('Life expectancy at birth (years)', 'GDP', data = zimbabwe_data, color = 'red')
plt.title('Zimbabwe')
plt.xlabel('Life Expectancy')
plt.ylabel('GDP')

plt.show()
plt.clf()

# GDP and life expectancy appear to generally have a positive correlation. 
```

```python
# Finding actual correlation values between life expectancy and GDP

chile_corr = chile_data['Life expectancy at birth (years)'].corr(chile_data['GDP'])
print("The correlation between life expectancy and GDP in Chile is " + str(round(chile_corr, 3)))

china_corr = china_data['Life expectancy at birth (years)'].corr(china_data['GDP'])
print("The correlation between life expectancy and GDP in China is " + str(round(china_corr, 3)))

germany_corr = germany_data['Life expectancy at birth (years)'].corr(germany_data['GDP'])
print("The correlation between life expectancy and GDP in Germany is " + str(round(germany_corr, 3)))

mexico_corr = mexico_data['Life expectancy at birth (years)'].corr(mexico_data['GDP'])
print("The correlation between life expectancy and GDP in Mexico is " + str(round(mexico_corr, 3)))

usa_corr = usa_data['Life expectancy at birth (years)'].corr(usa_data['GDP'])
print("The correlation between life expectancy and GDP in the USA is " + str(round(usa_corr, 3)))

zimbabwe_corr = zimbabwe_data['Life expectancy at birth (years)'].corr(zimbabwe_data['GDP'])
print("The correlation between life expectancy and GDP in Zimbabwe is " + str(round(zimbabwe_corr, 3)))
```
